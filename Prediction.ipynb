```python
import cv2,time
from cv2 import imread
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
```

```python
model = load_model('model.keras')
```

```python
val = ['normal driving',
      ' texting - right',
      'talking on the phone - right',
      'texting - left',
      'talking on the phone - left',
      'operating the radio',
      'drinking',
      'reaching behind',
      'hair and makeup',
      'talking to passenger'
      ]
```

```python
# Load the input video
cap = cv2.VideoCapture("op.mp4")

# Get the video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output_video.avi', fourcc, fps, (width, height))

# Process each frame of the input video
while True:
    ret, frame = cap.read()
    if ret:
        output = frame.copy()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (64, 64)).astype("float32")
        predict = model.predict((np.expand_dims(frame, axis=0) / 255 - 0.5))[0]
        
        # Get predicted classes based on threshold
        predicted_classes = [val[i] for i, p in enumerate(predict) if p > 0.5]
        if predicted_classes:
            predictedLabel = ", ".join(predicted_classes)
        else:
            predictedLabel = "Unknown"
        
        # Draw label on the frame
        cv2.putText(output, predictedLabel, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        # Write the frame to the output video
        out.write(output)
        
        # Display the frame
        cv2.imshow('frame', output)
        
        # Break the loop if 'q' is pressed
        if cv2.waitKey(20) & 0xFF == ord("q"):
            break
    else:
        break

# Release the VideoCapture and VideoWriter objects
cap.release()
out.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```

```python
# Load the input image
image = cv2.imread("test/img_1.jpg")

# Convert image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize the image
image_resized = cv2.resize(image_rgb, (64, 64)).astype("float32")

# Normalize the image
image_normalized = (image_resized / 255.0 - 0.5)

# Perform prediction
predict = model.predict(np.expand_dims(image_normalized, axis=0))

# Get predicted classes based on threshold
predicted_classes = [val[i] for i, p in enumerate(predict[0]) if p > 0.5]
if predicted_classes:
    predictedLabel = ", ".join(predicted_classes)
else:
    predictedLabel = "Unknown"

# Draw label on the image
cv2.putText(image, predictedLabel, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

# Save the annotated image
cv2.imwrite('output_image.jpg', image)

# Display the annotated image
cv2.imshow('annotated_image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
